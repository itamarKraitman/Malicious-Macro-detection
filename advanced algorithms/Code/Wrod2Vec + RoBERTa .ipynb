```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from transformers import AdamW
from gensim.models import Word2Vec
import numpy as np
import pandas as pd
from joblib import dump, load
from utils import RobertaClassifier, TextDataset, save_loader, RobertaClassifierTrainer

from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

import warnings
warnings.filterwarnings('ignore')
```

```python
train_loader1 = load('/content/drive/MyDrive/Colab Notebooks/Malicious Macro Detection/train_loader.joblib')
validation_loader1 = load('/content/drive/MyDrive/Colab Notebooks/Malicious Macro Detection/val_loader.joblib')
test_loader1 = load('/content/drive/MyDrive/Colab Notebooks/Malicious Macro Detection/test_loader.joblib')
```

```python
class datasetForRoberta(Dataset):
  def __init__(self, dataset):
    self.text = dataset['vba_code']
    self.label = dataset['label']

  def __len__(self):
    return len(self.text)

  def __getitem__(self, idx):
    return self.text[idx], self.label[idx]
```

```python
mapper = {
    'white' : 1,
    'mal' : 0
}

train_set = pd.read_csv('train_dataset.csv', encoding='utf-16le')[['vba_code', 'label']]
val_set = pd.read_csv('validation_dataset.csv', encoding='utf-16le')[['vba_code', 'label']]
test_set = pd.read_csv('test_dataset.csv', encoding='utf-16le')[['vba_code', 'label']]

train_set['label'] = train_set['label'].map(mapper)
val_set['label'] = val_set['label'].map(mapper)
test_set['label'] = test_set['label'].map(mapper)

train_set = datasetForRoberta(train_set)
val_set = datasetForRoberta(val_set)
test_set = datasetForRoberta(test_set)

train_loader = DataLoader(train_set, batch_size=32, shuffle=True)
validation_loader = DataLoader(val_set, batch_size=32)
test_loader = DataLoader(test_set, batch_size=32)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
model = RobertaClassifier().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = AdamW(model.parameters(), lr=0.0001, eps=1e-8)
lr = 0.0001

trainer = RobertaClassifierTrainer(train_loader, validation_loader, test_loader, optimizer, criterion)
trainer.train()
#train_loader, val_loader, test_loader, optimizer, criterion ,epochs=4, lr=0.0001, device=None
```

```python
test_loss, test_accuracy, precision, recall, f1_score, conf_matrix = trainer.test()
```
