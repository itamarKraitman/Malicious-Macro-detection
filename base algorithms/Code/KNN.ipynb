```python
from vba_pipeline import *
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from utils import save_loader

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import make_scorer, recall_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
from joblib import load

from vba_pipeline import create_pipeline, evaluate_model
```

```python
param_grid_knn = {
    'model__n_neighbors': [3, 5, 7, 9],
    'model__weights': ['uniform', 'distance'],
    'model__p': [1, 2]  # p=1 is equivalent to manhattan distance, p=2 is equivalent to euclidean distance
}

mapper = {
    'white' : 1,
    'mal' : 0
}
```

```python
train_set = pd.read_csv('train_dataset.csv', encoding='utf-16le')
val_set = pd.read_csv('validation_dataset.csv', encoding='utf-16le')
test_set = pd.read_csv('test_dataset.csv', encoding='utf-16le')
```

```python
x_train, y_train = train_set['vba_code'], train_set['label']
x_val, y_val = val_set['vba_code'], val_set['label']
x_test, y_test = test_set['vba_code'], test_set['label']

y_train = y_train.map(mapper)
y_val = y_val.map(mapper)
y_test = y_test.map(mapper)
```

```python
knn_pipeline = create_pipeline(KNeighborsClassifier(), TfidfVectorizer())
```

```python
scorer = make_scorer(recall_score, average='binary')

# grid_search_knn = GridSearchCV(estimator=knn_pipeline, param_grid=param_grid_knn, scoring=scorer, cv=5)
rand_search_knn = RandomizedSearchCV(knn_pipeline, param_grid_knn, cv= 5, scoring=scorer)

# grid_search_knn
rand_search_knn
```

```python
# grid_search_knn.fit(x_train, y_train)

# print("Best parameters found: ", grid_search_knn.best_params_)
# print("Best recall score: ", grid_search_knn.best_score_)

rand_search_knn.fit(x_train, y_train)

print("Best parameters found: ", rand_search_knn.best_params_)
print("Best recall score: ", rand_search_knn.best_score_)
```

```python
# best_params = grid_search_knn.best_params_
best_params = rand_search_knn.best_params_
knn_params = {k.replace('model__', ''): v for k, v in best_params.items()}
knn_model = KNeighborsClassifier(**knn_params)

x_train_tfidf_1000 = load('x_train_1000.joblib')

knn_model.fit(x_train_tfidf_1000, y_train)
```
